```python
import matplotlib.pyplot as plt
x=[5,6,9]
y=[5,7,9]
p=[10,20,25]
plt.title("a line graph")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x,y ,marker='*')
plt.plot(x,p,marker='*')
plt.show()
```

```python
import matplotlib.pyplot as plt
months=['jan','feb','mar','apl','may','jun']
sales=[89,34,56,78,23,36]
plt.title('a bar graph')
plt.xlabel('months')
plt.ylabel('product sales')
plt.bar(months,sales)
plt.show()
```

```python
import matplotlib.pyplot as plt
months=['jan','feb','mar','apl','may','jun']
sales=[89,34,56,78,23,36]
plt.title('a bar graph')
plt.xlabel('product sales')
plt.ylabel('months')
plt.barh(months,sales)
plt.show()
```

```python
import matplotlib.pyplot as plt
data=[34,56,23,45,57,89,35,26,89,39,87]
plt.xlabel("x")
plt.ylabel("y")
plt.title("histogram plot")
plt.hist(data,bins=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
students=["sanju","jyothi","dhanu"]
points=[45,67,89]
plt.pie(points,labels=students)
plt.axis("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt
students=["sanju","jyothi","dhanu"]
points=[45,67,89]
c=['y','r','b']
plt.pie(points,labels=students,colors=c,shadow=True,
explode=(0.005,0.005,0.005),autopct='%1.1f%%')
plt.axis('equal')
plt.legend()
plt.show()

```
